# **Day 4: Visualisation in Python**

Data visualisation is an essential part of data analysis. It helps transform raw data into insightful charts and graphs that are easier to interpret and communicate. Python offers a variety of powerful libraries that allow users to create both static and interactive visualisations with ease and flexibility.

In this notebook, we explore how to use Python's visualisation tools to gain insights from a real-world dataset that tracks the number of airline passengers over time. The goal is to help you understand how to visualise trends, comparisons, distributions, and relationships within the data using various types of charts.

### ðŸ§° Libraries Used for Visualisation

- **Matplotlib**  
  A foundational plotting library in Python that enables the creation of high-quality static, animated, and interactive visualisations. It offers fine-grained control over all elements of a plot.

- **Seaborn**  
  Built on top of Matplotlib, Seaborn makes it easier to create visually appealing and informative statistical graphics. It integrates well with pandas DataFrames and provides convenient functions for common plot types.

- **Plotly**  
  A modern graphing library that enables interactive visualisations. It allows users to zoom, pan, and hover over charts, making it ideal for web applications and dashboards.

- **NumPy**  
  Although not a visualisation library, NumPy is a fundamental package for numerical operations. It supports array-based computation and is used for trendline analysis in scatter plots.

Together, these libraries offer a complete toolkit for visualising data effectively in Python.

# Importing the required **libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns  # Needed for heatmaps
```

# Column **Chart**

```python
# Column chart
categories = ["A", "B", "C", "D"]
values = [10, 20, 15, 25]

plt.bar(categories, values, color="blue")
plt.title("Column Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

# Bar **Chart**

```python
# Column chart
plt.barh(categories, values, color="green")
plt.title("Column Chart Example")
plt.xlabel("Values")
plt.ylabel("Categories")
plt.show()
```

# Line **Chart**

```python
# Line chart
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 18, 25]

plt.plot(x, y, marker="o", linestyle="-", color="red")
plt.title("Line Chart Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# Scatter **Chart**

```python
# Scatter chart
x = np.random.rand(50) * 10
y = np.random.rand(50) * 10

plt.scatter(x, y, color="purple", alpha=0.6)
plt.title("Scatter Plot Example")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

### **Scatter Plot with a Trend Line** <br> np.polyfit() is a NumPy function used to find the best-fit polynomial equation for a given set of x and y data points.

```python
# Generate random scatter data
x = np.random.rand(50) * 10  # 50 random values between 0 and 10
y = 2 * x + np.random.randn(50) * 2  # Linear relationship with some noise

# Create scatter plot
plt.scatter(x, y, color="purple", alpha=0.6, label="Data Points")

# Compute the trend line (best-fit line)
m, b = np.polyfit(x, y, 1)  # Fit a linear regression line (degree=1)
plt.plot(x, m*x + b, color="red", linewidth=2, label="Trend Line")  # Plot trend line

# Add labels and title
plt.title("Scatter Plot with Trend Line")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.legend()

# Show the plot
plt.show()
```

# **Heat map**

```python
# Heatmap visualization
data = np.random.rand(5,5)  # Creating random data for the heatmap
sns.heatmap(data, annot=True, cmap="coolwarm")
plt.title("Heatmap Example")
plt.show()
```

# Stacked Column **Chart**

```python
# stacked bar chart
categories = ["A", "B", "C", "D"]
values1 = [5, 10, 15, 20]
values2 = [7, 8, 6, 10]

plt.bar(categories, values1, color="blue", label="Group 1")
plt.bar(categories, values2, bottom=values1, color="orange", label="Group 2")
plt.title("Stacked Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.legend()
plt.show()
```

# **Stacked Bar Chart**

```python
# Stacked column chart
plt.barh(categories, values1, color="blue", label="Group 1")
plt.barh(categories, values2, left=values1, color="orange", label="Group 2")
plt.title("Stacked Column Chart Example")
plt.xlabel("Values")
plt.ylabel("Categories")
plt.legend()
plt.show()
```

# Stacked Area **Chart**

```python
# Data
x = np.arange(1, 6)
y1 = np.array([10, 20, 15, 25, 30])
y2 = np.array([5, 10, 7, 12, 15])

# Plot stacked area chart
plt.fill_between(x, 0, y1, color="blue", alpha=0.6, label="Group 1")
plt.fill_between(x, y1, y1 + y2, color="orange", alpha=0.6, label="Group 2 (Stacked)")

# Labels and Title
plt.title("Stacked Area Chart Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)

# Show the plot
plt.show()


```

# Histogram **Visualisation**

```python
# Histogram visualization
data = np.random.randn(1000)  # Generate random data
plt.hist(data, bins=30, color="blue", alpha=0.7, edgecolor="black")
plt.title("Histogram Example")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

### **Creating a Simple Count Plot** <br> A Count Plot is a type of bar plot that shows the frequency of categorical values in a dataset. It counts how many times each category appears and displays the counts as bars.

```python
# Sample dataset
data = pd.DataFrame({
    "Category": ["A", "B", "A", "C", "B", "B", "A", "C", "C", "A"]
})

# FIX: Use `hue="Category"` and `legend=False`
sns.countplot(x="Category", hue="Category", data=data, palette="viridis", legend=False)

# Add labels and title
plt.xlabel("Category")
plt.ylabel("Count")
plt.title("Count Plot Example")

# Show the plot
plt.show()
```

**Count Plot with Hue (Grouping by Another Category)** <br> The hue parameter in Seaborn is used to differentiate data points by color based on a categorical variable.

```python
# Sample dataset with two categorical variables
data = pd.DataFrame({
    "Category": ["A", "B", "A", "C", "B", "B", "A", "C", "C", "A"],
    "Gender": ["M", "F", "F", "M", "M", "F", "F", "M", "F", "M"]
})

# Count plot grouped by 'Gender'
sns.countplot(x="Category", hue="Gender", data=data, palette="Set2")

# Labels and title
plt.xlabel("Category")
plt.ylabel("Count")
plt.title("Count Plot with Hue")

plt.show()
```

# Plotly **Demo**

```python
import plotly.express as px

# Sample dataset
data = pd.DataFrame({
    "Category": ["A", "B", "A", "C", "B", "B", "A", "C", "C", "A"],
    "Gender": ["M", "F", "F", "M", "M", "F", "F", "M", "F", "M"]
})

# Create a count plot using Plotly
fig = px.histogram(data, x="Category", color="Gender", barmode="group",
                   title="Category Count by Gender",
                   labels={"Category": "Category", "count": "Count"})

# Show the interactive plot
fig.show()
```
